## Questions

### Question 2:
#### Create a bar chart, histogram, scatterplot, and line chart

```python
# creating a bar chart
bar = ords_prods_cust['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green'])

x = ['Sat', 'Sun', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri']
default_x_ticks = range(len(x))
plt.plot(default_x_ticks)
plt.xticks(default_x_ticks, x)
plt.xticks(rotation = 360)

plt.xlabel("Day of Week", fontsize = 10),
plt.ylabel("Number of Orders in Millions", fontsize = 10),
plt.title("Number of Orders Per Day", fontsize = 15)
```

```python
# saving image
bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_orders_dow.png'))
```

```python
# creating histogram 
hist = ords_prods_cust['prices'].plot.hist(bins = 25)
plt.xlabel('Prices of Products', fontsize = 10),
plt.ylabel('Sum of Prices in Millions', fontsize = 10),
plt.title('Frequency of Prices', fontsize = 15)
```

```python
# saving image
hist.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'hist_prices.png'))
```

```python
# creating scatterplot
scatter = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_cust)
```

```python
# saving image
scatter.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatter.png'))
```

## Question 7:
#### Explore whether there's a connection between age and income using scatterplot

```python
# creating a scatterplot
scatter_age_income = sns.scatterplot(x = 'age', y = 'income', data = ords_prods_cust)
plt.xlabel('Age', fontsize = 10)
plt.ylabel('Income', fontsize = 10)
plt.title('Relationship Between Customers Age and Their Income', fontsize = 15)
```
