```python
# dropping 'unnamed' column
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# dropping unnamed column and _merge column
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['Unnamed: 0', '_merge'])
```
