#### Question 4: Create a loyalty flag for existing customers using transform() and loc()

```python
# creating "max_order" column
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head()
```

```python
# creating loyalty flag 
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged.head()
```

#### Question 5: Check the basic stats of the product prices for each loyalty category

```python
# performing multiple aggregations, grouping them by loyalty_flag
ords_prods_merged.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

#### Question 6: create a spending flag for each user based on the avg price across all their orders

```python
# creating 'mean_order' column
ords_prods_merged['mean_order'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merged.head()
```

```python
# creating spending flag
ords_prods_merged.loc[ords_prods_merged['mean_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['mean_order'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

#### Question 7: create an order frequency flag

```python
# creating median_order column
ords_prods_merged['median_order'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# creating order frequency flag
ords_prods_merged.loc[ords_prods_merged['median_order'] > 20, 'order_frequency_flag'] = "Non-frequent customer"
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_order'] > 10) & (ords_prods_merged['median_order'] <= 20), 'order_frequency_flag'] = "Regular customer"
```

```python
ords_prods_merged.loc[ords_prods_merged['median_order'] <= 10, 'order_frequency_flag'] = "Frequent customer"
```

```python
ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
