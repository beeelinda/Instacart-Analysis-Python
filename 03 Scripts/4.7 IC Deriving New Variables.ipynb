## 03. Questions

#### Question 1a: creating the price_label column

```python
df_ords_prods_merged.head()
```

```python
# creating the 'price_label' column
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_label'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_label'] = 'Mid-range product'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_label'] = 'Low-range product'
```

```python
df_ords_prods_merged['price_label'].value_counts(dropna = False)
```

#### Question 1b: creating the busiest_day column

```python
# finding the frequence of the orders_day_of_week column
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# creating a for-loop to make a new column for busiest days

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
# adding result to df
df_ords_prods_merged['busiest_day'] = result
```

```python
# printing the frequency of new column
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

#### Question 2: create a new column having the 2 busiest_days and 2 least_busy days

```python
# using for-loop to make a new column showing the busiest days and least busiest days

result = []

for day in df_ords_prods_merged["orders_day_of_week"]:
    if day == 0 or day == 1:
        result.append("Busiest days")
    elif day == 3 or day == 4:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
# adding the result to the df
df_ords_prods_merged['busiest_day'] = result
```

#### Question 4: create a new column "busiest_period_of_day"

```python
# finding out the frequency of the order_hour_of_day

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# creating a for-loop to make a new column for the "busiest_period_of_day"

result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
    if value == 9 or value == 10 or value == 11 or value == 12 or value == 13 or value == 14 or value == 15 or value == 16:
        result.append("Most orders")
    elif value == 7 or value == 8 or value == 17 or value == 18 or value == 19 or value == 20 or value == 21 or value == 22:
        result.append("Average orders")
    else:
        result.append("Fewest orders")
```

```python
result
```

```python
# adding result to df
df_ords_prods_merged['busiest_period_of_day'] = result
```
