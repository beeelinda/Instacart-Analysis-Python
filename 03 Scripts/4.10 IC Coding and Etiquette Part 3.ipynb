## 3. Question 6:
#### Create an appropriate visualization to show the distribution of profiles

```python
# creating distribution chart for ages
bar_age_group_distr = df_final['age_flag'].value_counts().sort_index().plot.bar(color = ['blue', 'purple', 'indigo'])
plt.xlabel('Age Group', fontsize = 10)
plt.ylabel('Frequency in Millions', fontsize = 10)
plt.xticks(rotation = 360)
plt.title('Distribution of Age Groups', fontsize = 15)
```

```python
# exporting distribution chart for age groups
bar_age_group_distr.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_age_group_distribution.png'))
```

```python
# creating distribution chart for income
bar_income_distr = df_final['income_flag'].value_counts().plot.bar(color = ['#11d2d2', '#0ca5a5', '#043333'])
plt.xlabel('Income', fontsize = 10)
plt.xticks(rotation = 360)
plt.ylabel('Frequency in Millions', fontsize = 10)
plt.title('Frequency of Customer Income')
```

```python
# exporting the distribution chart for income
bar_income_distr.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_income_distribution.png'))
```

```python
# creating distribution chart for family profile
bar_fam_profile_distr = df_final['family_profile'].value_counts().plot.bar(color = ['#ac33ff', '#7d33ff', '#5733ff'])
plt.xlabel('Family Status', fontsize = 10)
plt.xticks(rotation = 5)
plt.ylabel('Frequency in Millions', fontsize = 10)
plt.title('Distribution of Family Status', fontsize = 15)
```

```python
# exporting distribution chart for family profile
bar_fam_profile_distr.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_family_profile_distribution.png'))
```

```python
# creating department distribution chart
bar_department_distr = df_final['department'].value_counts().plot.barh()
plt.xlabel('Frequency', fontsize = 10)
plt.ylabel('Department', fontsize = 10)
plt.title('Distribution of Departments', fontsize = 15)
```

```python
# exporting department distribution chart
bar_department_distr.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_department_distribution.png'))
```

## 4. Question 7: 
#### Aggregate the max, mean, and min variables on a customer-profile level for usatge frequency and expenditure

```python
# aggregating prices and order number for age groups
df_final.groupby('age_flag').agg({'prices' :['mean', 'min', 'max'], 'order_number' :['mean', 'min', 'max']})
```

It looks like there's no significant difference between the age groups when it comes to prices and order numbers

```python
# aggregating prices and order number for income groups
df_final.groupby('income_flag').agg({'prices' :['mean', 'min', 'max'], 'order_number' : ['mean', 'min', 'max']})
```

Again it doesn't look like there's any significant difference between them

```python
# aggregating prices and order nubmers for family profiles
df_final.groupby('family_profile').agg({'prices' :['mean', 'min', 'max'], 'order_number' :['mean', 'min', 'max']})
```

Same here, not a lot of difference between the different groups

```python
# aggregating prices and order numbers for department
df_final.groupby('department').agg({'prices' :['mean', 'min', 'max'], 'order_number' :['mean', 'min', 'max']})
```

```python
# looking at the average prices for each department
df_final.groupby('department').prices.mean().sort_values()
```

```python
# creating a chart of the avg prices of departments
bar_avg_department_price = df_final['department'].value_counts().plot.barh()
plt.xlabel('Frequency', fontsize = 10)
plt.ylabel('Department', fontsize = 10)
plt.title('Average Distribution of Departments', fontsize = 15)
```

```python
# exporting chart
bar_avg_department_price.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_average_department_price.png'))
```
